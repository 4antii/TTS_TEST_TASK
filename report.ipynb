# Отчет по тестовому заданию 

## Какая задача решалась?

Ставилась задача обучения эмбединг модели для решения задачи верификации. 

## Основная идея метода? В чем отличие от других моделей?

Основная идея метода в следующем: на сырых аудио данных применяются параллельные сверточные слои с разными рецептивными полями, выход конкетинируется, прогоняется через 3 последовательных tf-SE слоя, потом все это конкетинируется со скипами и подается в TDNN агрегатор, на его выходе получаются эмбединги. Обучается через AM Softmax, по сути это должно увеличить расстояние между классами, что как раз важно для задачи классификации.

Отличие данной модели от других это обучение на сырых данных (raw audio), использование параллельного блока сверток с разными рецептивными полями, одной из реализаций механизма self-attention (tf-SE), TDNN. Множество других моделей используют mel-спектрограммы вместо необработанного аудио. 

## Какой эксперимент ставился? Какие получились данные и как их можно интерпритировать?

Ставился эксперимент с обучением модели на датасете VoxCeleb1, КОТОРЫЙ КСТАТИ УДАЛИЛИ с официального сайта и его теперь нельзя скачать, пришлось на это несколько дней потратить, чтобы найти его у знакомых. (Вопрос: почему на тестовое пропустили статью с удаленным датасетом?) 

Для трейна использовался dev датасет, для теста test датасет и файл с данными для верификации. 

Предобработка, как и в статье: load -> trancatuate -> normalise -> save_as_np (чтобы быстро загружать).

Модель обучалась примерно 35 часов на Tesla T4, AM-Softmax упал с 20.5 до 10.5

Основная метрика (Equal Error Rate): eer = 28.3775% threshold = 0.3013, что не является хорошим результатом. Скорее всего так получилось потому, что мне не хватило времени на то, чтобы хорошо обучить модель. Автор статьи обучил 300 эпох на Voxceleb2, который в 8 раз больше, чем VoxCeleb1, на котором я обучался примерно 50 эпох.

В текущем виде никак нельзя использовать, т.к слишком большая ошибка. К цифрам приблизиться не удалось, у меня ошибка в 10 раз больше. Перспективы развития заключаются в дообучении модели, после этого ее, вероятно, можно будет использовать для верефикации.

## Часть 2

Меня привлекает сфера моделирования анлогового аудио оборудования (у меня есть звукорежиссурский бэкграунд) с помощью нейронных сетей. Раньше это делалось с помощью линеаризации и решения системы диффуров в рил тайме, на математические расчеты уходило много времени. Сейчас же можно обучить модель, которая смоделирует все сама. 

Наиболее полезной мне показалась статья: Efficient Neural Networks for Real-time Analog Audio Effect Modeling (https://arxiv.org/pdf/2102.06200.pdf). В данной стаье хороший обзор моделей, много полезных лосс функций применительно к этой задаче. Даны понятные оценки на работу в режиме реального времени. 

Используя одну из конфигураций я сделал свою статью "Моделирование динамических характеристик аналоговых аудио компрессоров с использованием рекуррентных нейронных сетей", в которой показал новый подход к моделированию аудио компрессоров.
